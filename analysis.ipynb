```python
remove_col = df.dropna(subset=["show_id"],inplace=True)
print("Remove_column:",remove_col)
```

```python
df.drop_duplicates(inplace=True)
df
```

**Handle Missing Values**

```python
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
df["date_added"] = df["date_added"].fillna(df["release_year"].astype(str))
df['rating'] = df['rating'].fillna("No rating")
df['duration'] = df['duration'].fillna("0 min")
```

**Data Cleaning**


```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
cols = ["title", "director", "country", "listed_in"]
for col in cols:
    df[col] = df[col].str.strip()
```

**Duration Process**

```python
df["duration_num"] = df["duration"].str.extract(r'(\d+)').astype(float)
df["duration_num"] = pd.to_numeric(df["duration_num"], errors="coerce")
df["duration_type"] = df["duration"].str.extract(r'([a-zA-Z]+)')
df["duration_type"] = df["duration_type"].replace({
    "Season": "Seasons"
})
```

```python
if 'duration' in df.columns:
    df = df.drop("duration", axis=1)
df
```

**EDA & Visulization**


```python
#Univariate analysis
#Movie vs TV Show
import matplotlib.pyplot as plt

# Analysis
type_counts = df["type"].value_counts()
print(type_counts)

# Visualization
type_counts.plot(kind="bar")

plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")

plt.show()
print("Netflix has more Movies than TV Shows")
```

```python
country_counts = df["country"].value_counts().head(10)
print(country_counts)

country_counts.plot(kind="barh")

plt.title("Top 10 Countries")
plt.xlabel("Count")
plt.ylabel("Country")

plt.show()
print("Few countries dominate content production")
```

```python
#rating distribution
rating_counts = df["rating"].value_counts().sort_values(ascending=False).head()

rating_counts.plot(kind="bar")

plt.title("Rating Distribution (Sorted)")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

```python
release_year_counts = df['release_year'].value_counts().head(5)
print("release_year_counts:")
print(release_year_counts)
df['release_year'].value_counts().head(10).plot(kind='barh')
plt.show()
```

```python
year_counts = df["release_year"].value_counts().sort_index()
print(year_counts)

year_counts.plot()

plt.title("Content Released Over Years")
plt.xlabel("Year")
plt.ylabel("Count")

plt.show()
```

```python
type_duration = df.groupby("type")["duration_num"].mean()
print(type_duration)

type_duration.plot(kind="bar")

plt.title("Average Duration by Type")
plt.xlabel("Type")
plt.ylabel("Average Duration")

plt.show()
```

```python
movies = df[df["duration_type"] == "min"]

plt.scatter(movies["release_year"], movies["duration_num"])

plt.title("Release Year vs Movie Duration")
plt.xlabel("Release Year")
plt.ylabel("Duration (minutes)")

plt.show()
```
